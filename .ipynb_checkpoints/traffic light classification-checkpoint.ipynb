```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#importing the data
with open ('valid.p','rb') as valid:
    valid = pickle.load(valid)
    
with open ('train.p','rb') as train:
    train = pickle.load(train)
    
with open('test.p','rb') as test: 
    test = pickle.load(test)
```

```python
Xtrain, Ytrain = train['features'], train['labels']
Xvalid, Yvalid = valid['features'], valid['labels']
Xtest, Ytest = test['features'], test['features']
```

```python
#shuffling all the data present in the dataset
from sklearn.utils import shuffle
X_train, y_train = shuffle(Xtrain, Ytrain)
```

```python
#convert all the colorful image into black and white image
X_train_gray = np.sum(X_train/3, axis= 3, keepdims=True)
X_test_gray = np.sum(Xtest/3, axis = 3, keepdims = True)
X_val_gray = np.sum(Xvalid/3, axis =3 , keepdims = True)
```

```python
X_train_gray.shape,X_test_gray.shape,X_val_gray.shape
```

```python
i = 200
plt.imshow(X_train_gray[i].squeeze(), cmap='gray')
```

```python
#normalizing all the greyed imaged
x_train_gray_norm = (X_train_gray-128)/128
x_valid_gray_norm = (X_val_gray-128)/128
x_test_gray_norm = (X_test_gray-128)/128
```
